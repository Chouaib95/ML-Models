```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.cluster import KMeans
```

```python
from sklearn.datasets  import make_blobs
```

```python
data = make_blobs(n_samples = 300, n_features = 2, centers = 3, cluster_std = 1.5, 
                  random_state = 100)
```

```python
data
```

```python
data[0]
```

```python
data[0].shape
```

```python
plt.scatter(data[0][:,0], data[0][:,1])
```

```python
plt.scatter(data[0][:,0], data[0][:,1], c = data[1], cmap = 'viridis')
```

```python
dataMeans = KMeans(n_clusters = 6)
```

```python
dataMeans.fit(data[0])
```

```python
dataMeans.cluster_centers_
```

```python
dataMeans.labels_
```

```python
figure, (axe1, axe2) = plt.subplots(1, 2, sharey = 'row', figsize = (12,8))
axe1.set_title('representation des données après KMeans ')
axe1.scatter(data[0][:,0], data[0][:,1], c = dataMeans.labels_ , cmap = 'plasma' )

axe2.set_title('representation des données originales ')
axe2.scatter(data[0][:,0], data[0][:,1], c = data[1] , cmap = 'plasma' )
```
