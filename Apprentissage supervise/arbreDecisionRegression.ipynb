```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor, export_graphviz
```

```python
dataCarburant = pd.read_csv('dataCarburant.csv')
```

```python
dataCarburant
```

```python
dataCarburant.info()
```

```python
dataCarburant.describe()
```

```python
X = dataCarburant.drop('consommation', axis = 1)
y = dataCarburant['consommation']
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 100)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
model = DecisionTreeRegressor()
```

```python
model.fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)
```

```python
model.score(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
dataPrediction = pd.DataFrame({'Consommation réelle': y_test,
                               'Consommation prédite': predictions})
```

```python
dataPrediction
```

```python
export_graphviz(model, out_file = 'consommationCarburant.dot', 
               feature_names = X.columns, impurity = False, filled = True)
```
